## 5. Groupby & Aggregation `(13 pts)`

Compute the average Attack per primary type.

```python
# put your answer here

df.groupby('Type 1')['Attack'].mean()
```

Compute the maximum HP per generation.


```python
# put your answer here
df.groupby('Generation')['HP'].max()
```

Compute the total number of Pokémon per primary type.

```python
# put your answer here
df.groupby('Type 1')['Name'].size()
```

For each type, compute:

- mean Attack

- mean Defense

- mean Speed

```python
# put your answer here
mean_stats_by_type1 = df.groupby('Type 1')[['Attack', 'Defense', 'Speed']].mean()
mean_stats_by_type2 = df.groupby('Type 2')[['Attack', 'Defense', 'Speed']].mean()
# Displaying the result
print(mean_stats_by_type1)
print(mean_stats_by_type2)
```

For each generation, compute:

- count

- mean Total

- number of Legendary Pokémon (hint: use sum on Boolean)

```python
# put your answer here
grouped_by_gen = df.groupby('Generation')

# Calculating the count of Pokémon, mean of 'Total', and number of Legendary Pokémon for each generation
result = grouped_by_gen.agg(
    count=('Name', 'size'),
    mean_total=('Total', 'mean'),
    num_legendary=('Legendary', lambda x: x.sum())  # Sum of Boolean values (True = 1, False = 0)
)

print(result)
```

Which type combination (Type 1 + Type 2) has the highest average Attack?

```python
# put your answer here
# Combining Type 1 and Type 2 to create a unique Type combination
df['Type Combination'] = df['Type 1'] + " + " + df['Type 2']

# Grouping by 'Type Combination' and calculating the mean 'Attack' for each combination
mean_attack_by_combination = df.groupby('Type Combination')['Attack'].mean()

# Finding the type combination with the highest mean attack
highest_attack_combination = mean_attack_by_combination.idxmax()
highest_attack_value = mean_attack_by_combination.max()

print(f"The type combination with the highest average Attack is {highest_attack_combination} with an average Attack of {highest_attack_value:.2f}.")
```

Which generation has the highest proportion of Legendary Pokémon?

```python
# put your answer here
# Grouping by 'Generation' to count total Pokémon and Legendary Pokémon
gen_grouped = df.groupby('Generation').agg(
    total_pokemon=('Name', 'size'),
    legendary_pokemon=('Legendary', lambda x: x.sum())
)

# Calculating the proportion of Legendary Pokémon in each generation
gen_grouped['legendary_proportion'] = gen_grouped['legendary_pokemon'] / gen_grouped['total_pokemon']

# Finding the generation with the highest proportion of Legendary Pokémon
highest_proportion_gen = gen_grouped['legendary_proportion'].idxmax()
highest_proportion_value = gen_grouped['legendary_proportion'].max()

print(f"The generation with the highest proportion of Legendary Pokémon is Generation {highest_proportion_gen} with a proportion of {highest_proportion_value:.2f}.")
```


Which primary type has the largest variance in HP?

```python
# put your answer here
# Grouping by 'Primary Type' and calculating the variance of 'HP' for each primary type
hp_variance_by_type = df.groupby('Type 1')['HP'].var()

# Finding the primary type with the largest variance in HP
largest_variance_type = hp_variance_by_type.idxmax()
largest_variance_value = hp_variance_by_type.max()

print(f"The primary type with the largest variance in HP is {largest_variance_type} with a variance of {largest_variance_value:.2f}.")
```

Which primary type has the highest median Speed?

```python
# put your answer here
# Grouping by 'Primary Type' and calculating the median 'Speed' for each primary type
median_speed_by_type = df.groupby('Type 1')['Speed'].median()

# Finding the primary type with the highest median Speed
highest_median_speed_type = median_speed_by_type.idxmax()
highest_median_speed_value = median_speed_by_type.max()

print(f"The primary type with the highest median Speed is {highest_median_speed_type} with a median Speed of {highest_median_speed_value}.")
```

Group Pokémon by whether they are Legendary or not. Compare:

- mean Total

- mean Attack

- mean Defense

- mean Speed

```python
# put your answer here
# Grouping by 'Legendary' and calculating the mean for Total, Attack, Defense, and Speed
grouped_by_legendary = df.groupby('Legendary')[['Total', 'Attack', 'Defense', 'Speed']].mean()

# Displaying the result
print(grouped_by_legendary)
```

Show the top 5 strongest types by mean Total.

```python
# put your answer here

# Grouping by 'Primary Type' and calculating the mean 'Total' for each type
mean_total_by_type = df.groupby('Type 1')['Total'].mean()

# Sorting the values in descending order to get the top 5 strongest types
top_5_strongest_types = mean_total_by_type.sort_values(ascending=False).head(5)

# Displaying the result
print(top_5_strongest_types)
```

Rank generations by their average Attack.

```python
# put your answer here
# Grouping by 'Generation' and calculating the mean 'Attack' for each generation
mean_attack_by_gen = df.groupby('Generation')['Attack'].mean()

# Sorting the generations by mean 'Attack' in descending order
ranked_generations = mean_attack_by_gen.sort_values(ascending=False)

# Displaying the result
print(ranked_generations)
```

Show the top 10 fastest Pokémon using nlargest(10, "Speed").

```python
# put your answer here
# Using nlargest to get the top 10 fastest Pokémon based on the 'Speed' column
top_10_fastest_pokemon = df.nlargest(10, 'Speed')

# Displaying the result
print(top_10_fastest_pokemon[['Name', 'Speed']])
```
